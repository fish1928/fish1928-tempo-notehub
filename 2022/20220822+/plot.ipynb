```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import os
import json
```

```python
folder_data = 'datasource'
filename_2ds = 'esxdeploy_0812_processed_fingerprint_embedding2.json'
# filename_labels = 'esxdeploy_0812_processed_fingerprint_embedding762.dbscan.eps1'
filename_labels = 'esxdeploy_0812_processed_fingerprint_embedding2.dbscan.eps3'

path_file_2ds = os.path.join(folder_data, filename_2ds)
path_file_labels = os.path.join(folder_data, filename_labels)

with open(path_file_2ds, 'r') as file:
    dict_filename_2d = json.load(file)
# end

with open(path_file_labels, 'r') as file:
    dict_filename_label = json.load(file)
# end

dict_filename_dict_info = {}

for filename, point_2d in dict_filename_2d.items():
    dict_filename_dict_info[filename] = {
        'point': point_2d,
        'label': None
    }
# end

for filename, label in dict_filename_label.items():
    dict_filename_dict_info[filename]['label'] = label
# end

units_info = []
for filename, dict_detail in dict_filename_dict_info.items():
    point = dict_detail['point']
    label = dict_detail['label']
    units_info.append([filename, point, label])
#end
```

```python
xs = [unit[1][0] for unit in units_info]
ys = [unit[1][1] for unit in units_info]
labels = [unit[2] for unit in units_info]
size_label = len(set(labels))
```

```python
plt.figure(figsize=(26,26))
sns.scatterplot(
    x=xs,
    y=ys,
    hue=labels,
    palette=sns.color_palette("hls", size_label),
    legend="full",
    alpha=1
)
```
